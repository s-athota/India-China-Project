## China - After World War 2

Following the second world war, China immediately found itself in a civil war between the Nationalist KMT Party and Communist Party. In 1949, the Communist Party won under the leadership of Mao Zedong, with support from the Soviet Union. Early in the 1950's, China saw lots of resistance from what Mao called "counterrevolutionaries." As a direct consequence, Mao launched a campaign between 1950 and 1952 to suppress these enemies of the state, leading to the deaths of anywhere between 1 million and 3 million people. In 1951, Mao introduced the three-anti campaign, the antis being corruption, waste, and bureaucracy. This sought to rid the party anyone deemed unreliable. In 1952, the five-anti campaign was also implemented, with the antis being bribery, theft of state property, tax evasion, fraud, and theft of state intelligence. Both campaigns were used as platforms to kill anyone accused to being against the communist party

## India vs. China - Today

```python
GDP = pd.read_excel('World GDP Data.xls')
gdp_ic = GDP.loc[GDP['Country Name'].isin(['China', 'India'])]
```

```python
gdpict = gdp_ic.transpose()
gdpict = gdpict.iloc[4:]
# Dividing the values by a trillion for graphing
# After transposing the graph, the years became the index and random numbers were assigned as the names of the columns
gdpict[38] = gdpict[38] / 1000000000000
gdpict[107] = gdpict[107] / 1000000000000
```

```python
plt.plot(gdpict.index, gdpict[38], marker='None', linestyle='-', label='China')
plt.plot(gdpict.index, gdpict[107], marker='None', linestyle='-', label='India')
plt.xticks(rotation=90)
rcParams['figure.figsize'] = 25, 25
xlabel('Year')
ylabel('GDP ($T)')
legend(loc='best')
show()
```

## How does urbanization differ between the two nations in the last several decades?

```python
urban = pd.read_excel('Urbanization Values 1980 2019.xlsx')
urban['China'] = urban['China'] * 100
urban['India'] = urban['India'] * 100
urban[-5:]
#source = Knoema
```
